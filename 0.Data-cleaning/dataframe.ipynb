# Data cleaning


## In ra thông tin của dataframe
Câu lệnh: `df.info()`

```python
df.info()
```

Có thể thấy kiểu dữ liệu của `Average_Pulse` và `Max_Pulse` là object. Để tính toán, có thể chuyển sang dạng float64 bằng `.astype(kiểu dữ liệu đích)` <br>
`df['Average_Pulse'] = df['Average_Pulse'].astype(float)`

```python
df2 = df.copy()
df2["Average_Pulse"] = df2['Average_Pulse'].astype(float)
df2["Max_Pulse"] = df2["Max_Pulse"].astype(float)
df2
```

## Đếm số lượng giá trị null trong df
Câu lệnh `missing_value_counts = df.isnull().sum()` <br>
kết quả trả về là 1 series, có thể thao tác như 1 list

```python
missing_values_count = df.isnull().sum()
print(type(missing_values_count))
missing_values_count
```

### tính phần trăm lượng dữ liệu lỗi trong df

```python
total_cells = np.product(df.shape) # nhân các giá trị cột và hàng của df
total_missing = missing_values_count.sum()
print('Phần trăm data lỗi là ', total_missing/total_cells * 100)
```

## Xóa dòng/cột chứa dữ liệu không hợp lệ
Câu lệnh: `df.dropna(axis=0/1, inplace=True/False)`
Trong đó:
- axis:
    - 0: theo chiều ngang (trục Ox)
    - 1: theo chiều dọc (trục Oy)
- inplace:
    - 0: thao tác không làm thay đổi dataframe đang thực hiện mà chỉ trả về 1 dataframe khác
    - 1: thao tác làm thay đổi dataframe đang thực hiện

```python
df1 = df.copy()
df2 = df1.dropna(axis=0, inplace=False) # xóa dòng chứa NA, AF ở df1 nhưng trả về 1 df khác
print('df1 = ', df1, sep='\n')
print('-----------------------------------------')
print('df2 = ', df2, sep='\n')
```

vì `inplace=False` nên df1 không bị thay đổi

```python
df3 = df.copy()
df4 = df3.dropna(axis=1, inplace=True)
print(df4)
```

Vì cột nào cũng có dữ liệu lỗi nên tất cả đều bị xóa

## Điền các giá trị còn thiếu
`df = df.fillna(value)`

```python
# Điền giá trị 0
df5 = df.copy()
df5.fillna(0, inplace=True)
df5
```

```python
# Điền vào giá trị ngay sau nó trong cột/hàng, nếu là cuối cùng (cột/hàng) thì điền 0
# axis = 0 là theo hàng, = 1 là theo cột
df6 = df.copy()
df6 = df6.fillna(method='bfill', axis=1).fillna(0)
df6
```

####  Điền bằng giá trị có tần suất lớn nhất (mode) trong từng cột
`df = df.fillna(df['tên_cột'].mode()[0], inplace=True)`

```python
df7 = df.copy()
df7
```

```python
df7['Duration'].fillna(df7['Duration'].mode()[0], inplace=True)
df7['Average_Pulse'].fillna(df7['Average_Pulse'].mode()[0], inplace=True)
df7['Max_Pulse'].fillna(df7['Max_Pulse'].mode()[0], inplace=True)
df7['Calorie_Burnage'].fillna(df7['Calorie_Burnage'].mode()[0], inplace=True)
df7['Hours_Work'].fillna(df7['Hours_Work'].mode()[0], inplace=True)
df7['Hours_Sleep'].fillna(df7['Hours_Sleep'].mode()[0], inplace=True)
df7
```

### Điền bằng giá trị trung bình (mean) của cột
`df.fillna(df['tên_cột'].mean(), inplace=True)`

```python
df8 = df.copy()
df8['Duration'].fillna(df7['Duration'].mean(), inplace=True)
df8['Average_Pulse'].fillna(df7['Average_Pulse'].mean(), inplace=True)
df8['Max_Pulse'].fillna(df7['Max_Pulse'].mean(), inplace=True)
df8['Calorie_Burnage'].fillna(df7['Calorie_Burnage'].mean(), inplace=True)
df8['Hours_Work'].fillna(df7['Hours_Work'].mean(), inplace=True)
df8['Hours_Sleep'].fillna(df7['Hours_Sleep'].mean(), inplace=True)
df8
```
