# Chuẩn hóa dữ liệu theo min max
$$ y = \frac{x - min}{max - min}$$
Các giá trị sẽ trở thành giá trị từ $0 - 1$

```python
def convert(df_row):
    return (df_row - df_row.min()) /(df_row.max() - df_row.min())
```

```python
# Phương pháp trung bình
df7 = df2.copy()
df7['sepal.length'] = convert(df7['sepal.length'])
df7['sepal.width'] = convert(df7['sepal.width'])
df7['petal.length'] = convert(df7['petal.length'])
df7['petal.width'] = convert(df7['petal.width'])
df7
```

# 6. Xây dựng mô hình hồi quy tuyến tính dự đoán variety dựa theo giá trị khác

## Xây dựng tay
### Thêm cột 1 vào X_train

```python
df_reg = df10.copy()
len(df_reg)
```

```python
X, y = df_reg.iloc[:, :-1].values, df_reg.iloc[:, -1]
one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis=1)
train_length = int(0.8 * len(df_reg))
```

```python
X_train, y_train, X_test, y_test = Xbar[:train_length, :], np.array([y[:train_length]]).T, Xbar[train_length:, :], np.array([y[train_length:]]).T
```

```python
X_train
```

```python
print(y_train.shape)
```

```python
A = np.linalg.pinv(np.dot(X_train.T, X_train))
b = np.dot(X_train.T, y_train)
```

```python
w = np.dot(A, b)
w
```

```python
#w_0, w_1, w_2, w_3, w_4 = w[0][0], w[1][0], w[2][0], w[3][0], w[4][0]
```

```python
y0 = np.dot(X_test, w)
y_pred = np.round(y0)
# tính toán độ chính xác
acc = sum([1 if y_test[i] == y_pred[i] else 0 for i in range(len(y_test))])
print('Accurancy: ', acc/len(y_test))
```

### Thư viện scikit-learn

#### Sử dụng X_train có cột 1 ở đầu

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression(fit_intercept=False) #fit_
regr.fit(X_train, y_train)
print(regr.coef_)
y_pred = np.round(regr.predict(X_test))
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```
