```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('train_samsung.csv')
df_train.head()
```

```python
df_train.info(); df_train.describe()
```

```python
df1 = df_train.copy()
df1['X1'].fillna(df1['X1'].mode()[0], inplace=True)
df1['X2'].fillna(df1['X2'].mode()[0], inplace=True)
df1['X3'].fillna(df1['X3'].mode()[0], inplace=True)
df1['X4'].fillna(df1['X4'].mode()[0], inplace=True)
df1['X5'].fillna(df1['X5'].mode()[0], inplace=True)
df1['X6'].fillna(df1['X6'].mode()[0], inplace=True)
df1['X7'].fillna(df1['X7'].mode()[0], inplace=True)
df1['X8'].fillna(df1['X8'].mode()[0], inplace=True)
df1['X9'].fillna(df1['X9'].mode()[0], inplace=True)
df1['X10'].fillna(df1['X10'].mode()[0], inplace=True)
for i in df1.select_dtypes('object').columns:
    df1[i] = LabelEncoder().fit_transform(df1[i])
df1.info()
```

```python
X, y = df1.drop(['Class'], axis=1), df1['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred
```

```python
print('accuracy: ', accuracy_score(y_test, y_pred))
```

```python
def preprocess(df):
    df1 = df.copy()
    df1['X1'].fillna(df1['X1'].mode()[0], inplace=True)
    df1['X2'].fillna(df1['X2'].mode()[0], inplace=True)
    df1['X3'].fillna(df1['X3'].mode()[0], inplace=True)
    df1['X4'].fillna(df1['X4'].mode()[0], inplace=True)
    df1['X5'].fillna(df1['X5'].mode()[0], inplace=True)
    df1['X6'].fillna(df1['X6'].mode()[0], inplace=True)
    df1['X7'].fillna(df1['X7'].mode()[0], inplace=True)
    df1['X8'].fillna(df1['X8'].mode()[0], inplace=True)
    df1['X9'].fillna(df1['X9'].mode()[0], inplace=True)
    df1['X10'].fillna(df1['X10'].mode()[0], inplace=True)
    for i in df1.select_dtypes('object').columns:
        le = LabelEncoder()
        df1[i] = le.fit_transform(df1[i])
    df1['X3'] = df1['X3'].replace(to_replace='3+', value=4)
    return df1
```

```python
df_test = pd.read_csv('test_samsung_noclass.csv')
dtset = preprocess(df_test)
dtset.info()
```

```python
model = LogisticRegression(C=0.23)
model.fit(X, y)
y_0 = model.predict(dtset)
```
