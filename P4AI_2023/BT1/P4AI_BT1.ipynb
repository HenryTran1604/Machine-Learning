```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('P4AI_BT1.csv')
df
```

```python
df.info(), df.isna().sum()
```

## Điền giá trị bị thiếu của các trường có giá trị số theo phương pháp trung bình

```python
df1 = df.copy()
for i in df1.select_dtypes(include=['float64', 'int64']):
    df1[i].fillna(df[i].mean(), inplace=True)    
```

```python
for i in df.select_dtypes(include=['object']):
    df1[i].fillna(df[i].mode()[0], inplace=True) 
```

```python
df1.info()
```

```python
df2 = df1.copy()
df2[(df2['sepal.length'] > 5) & (df2['sepal.width'] > 3)]
```

# Chuẩn hóa dữ liệu theo min, max
$$ x' = \frac{x - min}{max - min} $$

```python
df3 = df2.copy()
for i in df3.select_dtypes(include=['float64', 'int64']):
    df3[i] = (df3[i] - df3[i].min())/(df3[i].max() - df3[i].min()) 
df3
```

```python
from sklearn.preprocessing import LabelEncoder
df3['variety'] = LabelEncoder().fit_transform(df3['variety'])
df3
```

# Xây dựng hồi quy tuyến tính

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df4 = df3.copy()
X = df4.drop(['variety'], axis=1)
y = df4['variety']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 40)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
reg.coef_, reg.intercept_
```

```python
y_pred = reg.predict(X_test)
y_labels = [round(i) for i in y_pred]
```

```python
from sklearn.metrics import accuracy_score
print('accuracy: ', 100 * accuracy_score(y_labels, y_test))
```

# Tự xây dựng model

```python
X1 = np.asarray(df4.drop(['variety'], axis=1))
y1 = np.asarray(df4['variety'])
X1 = np.hstack((np.ones((X1.shape[0], 1)), X1))
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.25, random_state = 40)
```

```python
A = np.dot(X_train1.T, X_train1)
b = np.dot(X_train1.T, y_train1)
w = np.dot(np.linalg.pinv(A), b)
w
```

```python
y_pred1 = np.round(np.dot(X_test1, w.T))
print(accuracy_score(y_pred1, y_test1))
```
